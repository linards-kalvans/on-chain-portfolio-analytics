```python
from web3 import Web3
import requests
from typing import Dict, List, Optional
import os
```

```python
ETHERSCAN_KEY = os.getenv("ETHERSCAN_API_KEY")
ETHERSCAN_API_URL="https://api.etherscan.io/api"
INFURA_KEY = os.getenv("INFURA_API_KEY")
INFURA_URL = f"https://mainnet.infura.io/v3/{INFURA_KEY}"
w3 = Web3(Web3.HTTPProvider(INFURA_URL))
```

```python
# url = "https://api.etherscan.io/api"
test_wallets = [
    "0xde0B295669a9FD93d5F28D9Ec85E40f4cb697BAe",  # Ethereum Foundation
    "0x47ac0Fb4F2D84898e4D9E7b4DaB3C24507a6D503",   # Large DeFi wallet
    "0xd8dA6BF26964aF9D7eEd9e03E53415D37aA96045", # Vitalik Buterin
    "0x742d35Cc6634C0532925a3b844Bc454e4438f44e" # Binance
]
params = {
    "module": "account",
    "action": "txlist",
    "address": test_wallets[3],
    "apikey": ETHERSCAN_KEY,
    "sort": "desc"
}
session = requests.Session()
response = session.get(ETHERSCAN_API_URL, params=params)
response.json().get('result', [])
```
